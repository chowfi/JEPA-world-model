```python
####### pairing (single timestep) ########

import numpy as np
import matplotlib.pyplot as plt

x = np.load('data/train/states.npy', mmap_mode='r')  # (N, 17, 2, 65, 65)

n = 0
t0 = 15

agent_c = 0
layout_c = 1

maze = x[n, 0, layout_c]
pos = x[n, t0, agent_c]

y, x_ = np.unravel_index(np.argmax(pos), pos.shape)

heat = pos.astype(float)
fig, ax = plt.subplots(figsize=(4, 4), dpi=150)

ax.imshow(maze, cmap="gray_r", interpolation="nearest")
ax.imshow(heat, cmap="Greys", alpha=0.35, interpolation="bilinear")

ax.plot([x_], [y], "o", color="#1f77ff", ms=4)

ax.set_axis_off()
plt.tight_layout(pad=0)
plt.show()
```
